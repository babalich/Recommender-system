### 1. В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной филтьтрации?

Гибридные системы используют не только матрицу взаимодействий user-item, но и внешние данные о пользователях и товарах. <br>
В КФ мы вычисляем эмбеддинги для пользователя и товара, в гибридных - сумму эмбеддингов пользователя, помноженных на фичи.

### 2. Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы

1. Новый пользователь в соц. сети еще не имеет взаимодействий с "товаром". Но первым делом заполняется анкета: возраст, пол, интересы, любимые фильмы, увлечения и т.д. Эти данные - наши фичи. <br>
2. Отличный пример в статье с датасетов CrossValidated. У нас не просто каждый товар имеет признаковое описание, но и пользователь. При таким раскладе холодный старт вообще не проблема (хотя по сути это тоже самое что и первый пример) <br>

#### Какие основные отличия предложенной системы от тех подходов, которые мы разбирали на семинарах?
На мой взгляд - главное отличие в многоуровневости. Мы решали задачу целиком. В статье идет послойное решение задачи. Сначала постепенно усложняющаяся последовательность фильтров, котороая позволяет более экономично отсеивать нерелевантные результаты. Затем - ранжирование оставшихся.<br>
При этом для поисковой выдачи используются не только парные признаки запрос-вакансия, но и парные признаки вакансия-резюме а так же явный признак (отклик на вакансию)

#### Какие проблемы могут возникнуть при выводе такой модели в продакшен?


У разработчиков список вакансий несбалансирован по условным классам (премиум, стандарт и т.д). Более сильный класс отнимал занимал топовые места, поэтому приходится искуственно разделять выдачу. Как именно - задача тестирования. <br>
Слишком сильная фильтрация может отрезать релевантные вакансии, слишком слабая - сделать время работы система слишком долгой <br>
Для понимания поезности в Production можно ориентироватся, например, на увеличение откликов. Для offline тестирования модели пришлось вручную размечать подходящие вакансия на части выборки - это трудоемкий процесс

#### 3) С какими моделями сравнивают LightFM? Опишите их основные идеи кратко

а) MF. Матричная факторизация user * item и сигмойдой.

б) LSI-LR. Модель на основе контента. Латентные признаки извлекаются из матрицы признаков(с помощью Latent Semantic Indexing), после чего итем представляется в качестве линейной комбинации данных эмбеддингов. Для каждого пользователя строятся модели логистической регрессии по сформированным данным. То есть упрощенно, мы формируем эмбеддинги из тэгов, по этим данным строим модели по каждому из пользователей. И предсказание делается, подавая пользователю линейную сумму эмбеддингов товара, для которого мы делаем предсказание. 

в) LSI-UP. Гибридная система. Каждый товар - это вектор. Берется каждый профиль пользователя, представляетсяв виде линейной комбинации "товара". К получившейся матрице применяется LSI, получаем латентное представление как юзера, так и товара

г) Различные варианты LightFM (только тэги, тэги и взаимодействия, тэги и признаки пользователя )

Интересен так же момент про использование в Lyst. Отмечена высокая важность правильной разметки товаров (максимально подробной, целый отдел профильных специалистов в моде занимается заполнением фичей), а так же фичи, которые получаются в результате работы сверточных сетей.
